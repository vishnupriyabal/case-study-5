# 3.Check whether all the zones generate the same amount of sales.

H0: Null Hypothesis states that all zones generate the same amount of sales.

H1: Alternative Hypothesis is that all zones are not generate same amount of sales

Level of Significance:0.05

```python
from scipy.stats import f_oneway
F_statistic, pvalue  = f_oneway(data['Zone - A'], data['Zone - B'],data['Zone - C'],data['Zone - D'])
print("The calculated F statistic is :",F_statistic)
print("The P-value is :" ,pvalue)

alpha = 0.05

if pvalue < alpha:
    print("We REJECT the Null Hypothesis")
elif pvalue > alpha:
    print("We're Unable to reject the Null Hypothesis")
```

P value is less than 0.05, we can reject the null hypothesis and conclude that the zones are not all same.
